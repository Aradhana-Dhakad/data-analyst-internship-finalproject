# Data Cleaning

create  a copy of data for manupulation and processing

```python
var=data.copy()
```

```python
var
```

Total Charges is in Object datatype.Let's convert it into numerical datatype

```python
var.TotalCharges = pd.to_numeric(var.TotalCharges,errors='coerce')
 # if the error argument is passes as 'coerce',then invalid parsing will be set as NaN
var.isnull().sum()
```

As we can see there are 11 missing values in TotalCharges column.Let's check these records.

```python
var.loc[var["TotalCharges"].isnull()==True]
```

Since the percent of these records compared to total dataset is very low i.e. 0.15% , It is safe to ignore them for further processing

a=(11/7043)*100=0.15%



```python
# Removing Missing Values

var.dropna(how='any',inplace=True)
```

```python
var
```

Divide customer into bins based on tenure e.g. for tenure<12 months : assign a tenure group if 1-12, for tenure between 1 to 2 years, tenure group of 13-24; and so on 

```python
# Get the max tenure
print(var['tenure'].max())
```

```python
# Group the tenure in bins of 12 months
labels=["{0}-{1}".format(i,i+11) for i in range(1,72,12)]

var['tenure_group']=pd.cut(var.tenure,range(1,80,12),right=False , labels=labels)
```

```python
var['tenure_group'].value_counts()
```

Remove columns not required for Processing

```python
var.drop(columns=['customerID','tenure'],axis=1,inplace=True)
```

The get_dummies() function creates dummy variables from categorical variables.It takes 1 when the original categorical var was that level otherwise 0

```python
plt.figure(figsize=(20,8))
var_dummies.corr()['Churn'].sort_values(ascending=False).plot(kind='bar')
```
